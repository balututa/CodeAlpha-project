```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
import pandas as pd

# Correct function to read an Excel file
data = pd.read_excel('credit-data-training.xlsx')
```

```python
data
```

```python
data['Credit-Application-Result'].unique()
```

```python
label_mapping = {'Creditworthy':1, 'Non-Creditworthy':0}
data['Credit-Application-Result'] = data['Credit-Application-Result'].map(label_mapping)
```

```python
data
```

```python
data['Account-Balance'].unique()
```

```python
label_mapping = {'Some Balance':1, 'No Account':0}
data['Account-Balance'] = data['Account-Balance'].map(label_mapping)
```

```python
data['Payment-Status-of-Previous-Credit'].unique()
```

```python
label_mapping = {'Paid Up':1, 'No Problems (in this bank)':2,'Some Problems':0 }
data['Payment-Status-of-Previous-Credit'] = data['Payment-Status-of-Previous-Credit'].map(label_mapping)
```

```python
data['Purpose'].unique()
```

```python
label_mapping = {'Home Related':1, 'Used car':2,'Other':0,'New car':3  }
data['Purpose'] = data['Purpose'].map(label_mapping)
```

```python
data.info()
```

```python
data['Value-Savings-Stocks'].unique()
```

```python
label_mapping = {'< £100':1, '£100-£1000':2,'nan':0  }
data['Value-Savings-Stocks'] = data['Value-Savings-Stocks'].map(label_mapping)
```

```python
import numpy as np
data['Value-Savings-Stocks'] = data['Value-Savings-Stocks'].replace([np.inf, -np.inf], np.nan)  # Replace inf/-inf with NaN
data['Value-Savings-Stocks'] = data['Value-Savings-Stocks'].fillna(0).astype(int)  # Replace NaN with 0 and convert
```

```python
data['Length-of-current-employment'].unique()
```

```python
label_mapping = {'< 1yr':1, '1-4 yrs':2,'4-7 yrs':3 }
data['Length-of-current-employment'] = data['Length-of-current-employment'].map(label_mapping)
```

```python
data['Guarantors'].unique()
```

```python
label_mapping = {'nan':0, 'Yes':2}
data['Guarantors'] = data['Guarantors'].map(label_mapping)
```

```python
import numpy as np
data['Guarantors'] = data['Guarantors'].replace([np.inf, -np.inf], np.nan)  # Replace inf/-inf with NaN
data['Guarantors'] = data['Guarantors'].fillna(0).astype(int)  # Replace NaN with 0 and convert
```

```python
data['Duration-in-Current-address'].unique()
```

```python
import numpy as np
data['Duration-in-Current-address'] = data['Duration-in-Current-address'].replace([np.inf, -np.inf], np.nan)  # Replace inf/-inf with NaN
data['Duration-in-Current-address'] = data['Duration-in-Current-address'].fillna(0).astype(int)  # Replace NaN with 0 and convert
```

```python
data.drop(columns=['Length-of-current-employment'],inplace = True)
```

```python
import numpy as np
data['Age-years'] = data['Age-years'].replace([np.inf, -np.inf], np.nan)  # Replace inf/-inf with NaN
data['Age-years'] = data['Age-years'].fillna(0).astype(int)  # Replace NaN with 0 and convert
```

```python
data['Concurrent-Credits'].unique()
```

```python
data.drop(columns = ['Concurrent-Credits'],inplace = True)
```

```python
data['No-of-Credits-at-this-Bank'].unique()
```

```python
label_mapping = {'1':1, 'More than 1':2 }
data['No-of-Credits-at-this-Bank'] = data['No-of-Credits-at-this-Bank'].map(label_mapping)
```

```python
x = data.drop(['Credit-Application-Result'],axis = 1)
```

```python
x
```

```python
y = data['Credit-Application-Result']
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression()  # Create a logistic regression model
model.fit(X_train, y_train)   # Train the model
```

```python
y_pred = model.predict(X_test)  # Predict on test data
```

```python
# Accuracy Score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
